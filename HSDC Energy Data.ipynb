```python
# comparing the effects of Regularisation
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns= ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_weights = get_weights_df(model, X_train, "Linear Model Weights")
Ridge_weights = get_weights_df(ridge_reg, X_train, "Ridge Model Weights")
lasso_weights = get_weights_df(lasso_reg, X_train, "Lasso Model Weights")

final_weights = pd.merge(linear_weights, Ridge_weights, on= 'Features')
final_weights = pd.merge(final_weights, lasso_weights, on= 'Features')
```
